```python
import numpy as np
'''
Author: LawsonAbs
Date: 2021-09-16 18:53:02
LastEditTime: 2021-09-16 19:02:08
FilePath: /daguan/risk_data_grand/utils/tests.py
仔细分析每个文件在每个类别下的输出值
'''

import os
import matplotlib.pyplot as plt
from collections import Counter
def statistic(submission_path):
    id_label = {}
    with open(submission_path,'r') as f:
        f.readline()
        for line in f:
            iid,label = line.strip("\n").split(",")
            id_label[iid] = label
    cnt = Counter()
    for label in id_label.values():
        
        cnt[label]+=1
    
    print(cnt)
    return cnt



if __name__ == "__main__":
    x = [0,1]
    cnt = dict(statistic("/home/lawson/program/daguan/submission_balance_10_num_30.csv"))
    # cnt2 = dict(statistic("/home/lawson/program/daguan/submission_ensemble.csv"))
    # print(type(cnt))
    x = list(cnt.keys())
    index = np.arange(len(x))
    
    y = list(cnt.values())
    # y2 = list(cnt2.values())
    print(y)
    # print(y2)
    plt.figure(figsize=(30,10))
    for a,b in zip(x,y):    
        plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=17)        
    
    # for a,b in zip(x,y2):
    #     plt.text(x, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=17)        
    plt.tick_params(labelsize=15)
    plt.xlabel("class",fontdict={"fontsize":17})
    plt.ylabel("num",fontdict={"fontsize":17})
    plt.bar(x,y,width=0.3,align='center') # 画柱状图
    # plt.bar(index+0.4,y2,width=0.3,align='center') # 画柱状图
    plt.show()
```

```python
# 分析train+test中每行文本的长度
import matplotlib.pyplot as plt
def show_text_length(data_path):
    text_length = []
    with open(data_path,'r') as f:
        for line in f:
            line = line.strip("\n").split() # 以空格分
            text_length.append(len(line))
    plt.hist(text_length,bins=40,facecolor="blue",edgecolor="white",alpha=0.9)
data_path = "/home/lawson/program/daguan/risk_data_grand/data/all.txt"
show_text_length(data_path)
```
