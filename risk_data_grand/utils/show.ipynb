```python
'''
Author: LawsonAbs
Date: 2021-09-16 18:53:02
LastEditTime: 2021-09-16 19:02:08
FilePath: /daguan/risk_data_grand/utils/tests.py
仔细分析每个文件在每个类别下的输出值
'''

import os
import matplotlib.pyplot as plt
from collections import Counter
def statistic(submission_path):
    id_label = {}
    with open(submission_path,'r') as f:
        f.readline()
        for line in f:
            iid,label = line.strip("\n").split(",")
            id_label[iid] = label
    cnt = Counter()
    for label in id_label.values():
        
        cnt[label]+=1
    
    print(cnt)
    return cnt



if __name__ == "__main__":
    x = [0,1]
    cnt = dict(statistic("/home/lawson/program/daguan/submission_epoch_10_fl.csv"))
    # print(type(cnt))
    x = list(cnt.keys())
    y = list(cnt.values())
    plt.figure(figsize=(30,10))
    for a,b in zip(x,y):    
        plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=17)
    plt.xlabel("class",fontdict={"fontsize":17})
    plt.ylabel("num",fontdict={"fontsize":17})
    plt.bar(x,y,width=0.3,align='center') # 画柱状图
    
    plt.show()
    
```
